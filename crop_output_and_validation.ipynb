```python
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files

# === Step 1: Read AquaCrop output ===
crop_columns = [
    'Day','Month','Year','DAP','Stage','GD','Z','StExp','StSto','StSen','StSalt','StWeed',
    'CC','CCw','StTr','Kc(Tr)','Trx','Tr','TrW','Tr/Trx','WP','Biomass','HI',
    'Ydry','Yfresh','Brelative','WPet','Bin','Bout'
]
file_path = "ValidCrop.OUT"
crop_df = pd.read_csv(file_path, sep=r"\s+", header=None, names=crop_columns, encoding='latin1', skiprows=7)

# Convert to numeric
for col in crop_columns:
    crop_df[col] = pd.to_numeric(crop_df[col], errors="coerce")

# === Step 2: Define a plotting function ===
def plot_component(x, y, ylabel, title, filename, color='blue', linestyle='-'):
    fig, ax = plt.subplots(figsize=(16,10))
    ax.plot(crop_df[x], crop_df[y],
            linewidth=2, label=y,
            color=color, linestyle=linestyle)
    ax.set_xlabel("Days After Planting (DAP)", fontsize=24)
    ax.set_ylabel(ylabel, fontsize=24)
    ax.tick_params(axis='y', labelsize=20)
    ax.tick_params(axis='x', labelsize=20)
    ax.legend(fontsize=20)
    plt.title(title, fontsize=22)
    plt.grid(True)
    plt.savefig(filename, dpi=300, bbox_inches="tight")
    plt.close()
    files.download(filename)

# === Step 3: Generate plots ===
plot_component('DAP','CC',
               ylabel="Canopy Cover (%)",
               title="AquaCrop Canopy Cover",
               filename="Canopy_Cover.png",
               color='green')

plot_component('DAP','Biomass',
               ylabel="Biomass (ton/ha)",
               title="AquaCrop Biomass",
               filename="Biomass.png",
               color='orange')

plot_component('DAP','Ydry',
               ylabel="Yield (ton/ha)",
               title="AquaCrop Dry Yield",
               filename="Yield.png",
               color='purple')

plot_component('DAP','WPet',
               ylabel="Water Productivity (kg/m³)",
               title="AquaCrop Water Productivity",
               filename="Water_Productivity.png",
               color='red')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files
import numpy as np

# === Step 1: Read AquaCrop comparison output ===
comp_columns = [
    'Day','Month','Year','DAP','Stage',
    'CCsim','CCobs','CCstd',
    'Bsim','Bobs','Bstd',
    'SWCsim','SWCobs','SWstd',
    'Extra' # Added an extra column to handle ParserError: saw 15 instead of 14 fields
]

file_path = "ValidEvalData.OUT"

# Use engine='python' to handle irregular whitespace and extra columns
comp_df = pd.read_csv(file_path, sep=r"\s+", header=None,
                      names=comp_columns, encoding="latin1",
                      engine="python", skiprows=15)

# Convert numeric columns
for col in comp_columns:
    comp_df[col] = pd.to_numeric(comp_df[col], errors="coerce") # Corrected typo: comp[col] to comp_df[col]

# Replace AquaCrop missing value code (-9) with NaN
comp_df.replace(-9, np.nan, inplace=True)
comp_df.replace(-9.0, np.nan, inplace=True)

# === Step 2: Plot function for sim vs obs with error bars ===
def plot_sim_obs(x, ysim, yobs, ystd, ylabel, title, filename,
                 color_sim='blue', color_obs='red'):
    fig, ax = plt.subplots(figsize=(16,10))
    ax.plot(comp_df[x], comp_df[ysim],
            linewidth=2, label=f"{ysim}", color=color_sim)
    ax.errorbar(comp_df[x], comp_df[yobs],
                yerr=comp_df[ystd],
                fmt='o', color=color_obs, label=f"{yobs} ± std")
    ax.set_xlabel("Days After Planting (DAP)", fontsize=24)
    ax.set_ylabel(ylabel, fontsize=24)
    ax.tick_params(axis='y', labelsize=20)
    ax.tick_params(axis='x', labelsize=20)
    ax.legend(fontsize=20)
    plt.title(title, fontsize=22)
    plt.grid(True)
    plt.savefig(filename, dpi=300, bbox_inches="tight")
    plt.close()
    files.download(filename)

# === Step 3: Generate plots ===
plot_sim_obs('DAP','CCsim','CCobs','CCstd',
             ylabel="Canopy Cover (%)",
             title="Simulated vs Observed Canopy Cover",
             filename="CC_Sim_vs_Obs.png",
             color_sim='green', color_obs='black')

plot_sim_obs('DAP','Bsim','Bobs','Bstd',
             ylabel="Biomass (ton/ha)",
             title="Simulated vs Observed Biomass",
             filename="Biomass_Sim_vs_Obs.png",
             color_sim='orange', color_obs='purple')

plot_sim_obs('DAP','SWCsim','SWCobs','SWstd',
             ylabel="Soil Water Content (mm)",
             title="Simulated vs Observed SWC",
             filename="SWC_Sim_vs_Obs.png",
             color_sim='blue', color_obs='red')
```
